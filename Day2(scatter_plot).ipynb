Simple Scatter Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, '-o', color='black');
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x, y, color='black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
            label="maker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)

rng = np.random.RandomState(0)
for marker in ['o','.',',','d','x','<','>','s','D']:
  plt.scatter(rng.rand(5),rng.rand(5),label="marker='{0}'".format(marker))
  plt.legend(numpoints=1,bbox_to_anchor=(0.5,0.5,0.5,0.5),loc='upper right')
  plt.xlim(0,1.5)
```

```python
rnd = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')#alpha used to reduce or increase Transparensy
plt.colorbar();#show color  scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
